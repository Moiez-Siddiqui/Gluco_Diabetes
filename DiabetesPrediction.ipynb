# **Diabetes Prediction Using Machine Learning**

## **Introduction**
Diabetes is a collection of metabolic conditions characterized by elevated blood glucose levels over an extended period. Symptoms of high blood sugar include excessive urination, increased thirst, and heightened appetite. If left unmanaged, diabetes can lead to numerous health issues. Immediate complications can include diabetic acidosis, hyperosmolar hyperglycemic syndrome, or even fatality. Significant long-term consequences may involve heart disease, stroke, chronic kidney damage, foot sores, and visual impairment.

## **5) Training Testing Dataset**

```python
X=df.drop(['Outcome'],axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()

rf.fit(X_train,y_train)

y_pred = rf.predict(X_test)

accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)

accuracy_score(y_test,y_pred)
```

## **6) Predictive System**

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Assuming you already have your training data loaded as X_train and y_train
# X_train = ... (your training features)
# y_train = ... (your training labels)

# Example: Initialize and fit the scaler on training data
scaler = StandardScaler()
scaler.fit(X_train)

# Prompt the user for input
print("Please enter the following details:")
pregnancies = int(input("Number of pregnancies: "))
glucose = float(input("Glucose level: "))
blood_pressure = float(input("Blood pressure: "))
skin_thickness = float(input("Skin thickness: "))
insulin = float(input("Insulin level: "))
bmi = float(input("Body Mass Index (BMI): "))
diabetes_pedigree = float(input("Diabetes pedigree function: "))
age = int(input("Age: "))

# Create the input data tuple
input_data = (pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, diabetes_pedigree, age)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the input data
std_data = scaler.transform(input_data_reshaped)
print("Standardized Data:", std_data)

# Initialize and fit the classifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)

# Make the prediction
prediction = classifier.predict(std_data)
print("Prediction:", prediction)

# Output the result
if prediction[0] == 0:
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
